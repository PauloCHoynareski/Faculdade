```python
def decimal_para_binario(decimal):
    binario = ""
    while decimal > 0:
        binario = str(decimal % 2) + binario
        decimal = decimal // 2
    return binario

def decimal_para_octal(decimal):
    octal = ""
    while decimal > 0:
        octal = str(decimal % 8) + octal
        decimal = decimal // 8
    return octal

def decimal_para_hexadecimal(decimal):
    hexadecimal = ""
    while decimal > 0:
        resto = decimal % 16
        if resto < 10:
            hexadecimal = str(resto) + hexadecimal
        else:
            hexadecimal = chr(65 + resto - 10) + hexadecimal
        decimal = decimal // 16
    return hexadecimal

def binario_para_decimal(binario):
    decimal = 0
    for digito in binario:
        decimal = decimal * 2 + int(digito)
    return decimal

def octal_para_decimal(octal):
    decimal = 0
    for digito in octal:
        decimal = decimal * 8 + int(digito)
    return decimal

def hexadecimal_para_decimal(hexadecimal):
    decimal = 0
    for digito in hexadecimal:
        if digito.isdigit():
            decimal = decimal * 16 + int(digito)
        else:
            decimal = decimal * 16 + ord(digito.upper()) - 55
    return decimal

def converter_numero():
    print("Escolha a base de entrada:")
    print("1. Decimal")
    print("2. Binário")
    print("3. Octal")
    print("4. Hexadecimal")
    base_entrada = int(input("Digite o número correspondente à base: "))

    if base_entrada < 1 or base_entrada > 4:
        print("Erro: Número de base inválido.")
        return

    print("Escolha a base de saída:")
    print("1. Decimal")
    print("2. Binário")
    print("3. Octal")
    print("4. Hexadecimal")
    base_saida = int(input("Digite o número correspondente à base: "))

    if base_saida < 1 or base_saida > 4:
        print("Erro: Número de base inválido.")
        return

    numero = input("Digite o número a ser convertido na base de entrada escolhida: ")

    if base_entrada == 1:
        numero_decimal = int(numero)
    elif base_entrada == 2:
        numero_decimal = binario_para_decimal(numero)
    elif base_entrada == 3:
        numero_decimal = octal_para_decimal(numero)
    elif base_entrada == 4:
        numero_decimal = hexadecimal_para_decimal(numero)
    else:
        print("Erro: Número de base inválido.")
        return

    if base_saida == 1:
        resultado = numero_decimal
    elif base_saida == 2:
        resultado = decimal_para_binario(numero_decimal)
    elif base_saida == 3:
        resultado = decimal_para_octal(numero_decimal)
    elif base_saida == 4:
        resultado = decimal_para_hexadecimal(numero_decimal)
    else:
        print("Erro: Número de base inválido.")
        return

    print("O número convertido é:", resultado, "na base", base_saida)

converter_numero()
```
